```
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.ylabel('numbers')
plt.show()
```

```
plt.plot([2,4,6,8,10], [3,5,7,9,110])
plt.show()
```

```
plt.plot([1,2,3,4], [1,4,9,16],'ro')        # 'ro' 는 그래프에서 선대신 점을 찍어주는 역할이다.
#axis는 ([xmin, xmax, ymin, ymax]) x,y축 범위를 정해주는 함수다.
plt.axis([0,6,0,20])
plt.show()
```

```
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r+', t, t//t, t, t**2, 'g^')
plt.show()
```
