```
import tensorflow as tf
import pandas as pd

link = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
file = pd.read_csv(link)
# file.head()

# file.columns
IV = file[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat']]
SV = file[['medv']]
# print(IV.shape, SV.shape)

X = tf.keras.layers.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss = 'mse')

model.fit(IV, SV, epochs = 1000, verbose=0)
model.fit(IV, SV, epochs = 10)
```
